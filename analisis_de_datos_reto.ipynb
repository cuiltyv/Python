```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import os
```

```python
ruta = '../../../Downloads/data-LDM-ver1.0-2023-11-27.xlsx'

df = pd.read_excel(ruta)
#print("¿Existe el archivo?", os.path.exists(ruta))
```

```python
# Resumen rápido
print(df.describe())
print(df.info())
```

```python
# Visualizar distribución del GPA
sns.histplot(df['student.term_gpa'])
plt.title('Distribución del GPA por semestre')
plt.show()
```

```python
# Visualización
plt.figure(figsize=(10, 6))
plt.scatter(df['student_admission_test.score'], df['student.term_gpa_program'], alpha=0.7, color='royalblue')
plt.title('Relación entre Score de Admisión y GPA Final')
plt.xlabel('Score de Admisión')
plt.ylabel('GPA Final')
plt.grid(True)
plt.show()
```

```python
# Filtrar los datos para el rango de score de 1350 a 1600, el rango del PPA
df_filtered = df[(df['student_admission_test.score'] >= 1350) & (df['student_admission_test.score'] <= 1600)]

# Visualización
plt.figure(figsize=(10, 6))
plt.scatter(df_filtered['student_admission_test.score'], df_filtered['student.term_gpa_program'])
plt.title('Relación entre Score de Admisión (PAA) y GPA Final')
plt.xlabel('Score de Admisión')
plt.ylabel('GPA Final')
plt.grid(True)
plt.show()
```
