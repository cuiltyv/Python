# Clasificador Bayesiano Inegnuo

El clasificador bayesiano es un clasificador el se basa en el teorema de bayes, se basa en el principio de que la probabilidad de un evento se actualiza basada en nueva informacion. Por medio de la siguiente formula: La probabilidad P(Aâˆ£B), se puede calcular a partir de P(B|A) y las probabilidades marginales P(A) y P(B)

```python
import pandas as pd
from sklearn.naive_bayes import GaussianNB
from sklearn.metrics import confusion_matrix
from sklearn.metrics import accuracy_score
```

```python
train = pd.read_csv("mnist/mnist_train.csv")
test = pd.read_csv("mnist/mnist_test.csv")

X_train = train.drop("label", axis=1)
y_train = train["label"]

X_test = test.drop("label", axis=1)
y_test = test["label"]
```

```python
model_bi = GaussianNB()
model_bi.fit(X_train, y_train)

y_pred = model_bi.predict(X_test)

```

```python
#Matriz de confusion clasificador bayesiano
confusion_matrix(y_test, y_pred)
```

```python
#Accuracy clasificador bayesiano
accuracy_score(y_test, y_pred)
```

# Maquinas de vectores de soporte

Lo que hacen las maquinas de soporte de vectores es buscar separar los elementos por clase de la mejor manera posible.

```python
from sklearn import svm
```

```python
model_svm = svm.LinearSVC().fit(X_train, y_train)

y_pred = model_svm.predict(X_test)
```

```python
confusion_matrix(y_test, y_pred)
```

```python
accuracy_score(y_test, y_pred)
```

# Arboles de decision

El arbol de decision es una estructura en forma de arbol en el cual cada nodo representa una variable o caracteristica la cual dependiendo de la regla va a la siguiente rama o no y dependiendo donde acaben hay un label resultante

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.tree import plot_tree
```

```python
model_tree = DecisionTreeClassifier(random_state = 1025)
model_tree.fit(X_train, y_train)

y_pred = model_tree.predict(X_test)
```

```python
confusion_matrix(y_test, y_pred)
```

```python
accuracy_score(y_test, y_pred)
```

```python
print(plot_tree(model_tree))
```
